```python
df
```

```python
X=df.iloc[:,0:-1]
y=df['BurnoutRisk']
```

```python
from sklearn.feature_selection import SelectKBest,mutual_info_classif

mic=SelectKBest(score_func=mutual_info_classif,k=3)
```
