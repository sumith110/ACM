```python


df['work_hours'] =  (120- (df['SleepHours'] + df['PhysicalActivityHrs'] + df['CommuteTime']))/5
```

```python

medianworkhours = df['work_hours'].median()


df.loc[df['work_hours'] > 18, 'work_hours'] = medianworkhours
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='SleepHours', y='StressLevel', hue='BurnoutRisk')
plt.title('Sleep Hours vs Stress Level')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='HasMentalHealthSupport', hue='BurnoutRisk')
plt.title('Mental Health Support vs Burnout Risk')
plt.xticks(rotation=45)
plt.show()
```
