```python
df.drop('StressLevel',axis=1,inplace=True)
```

```python
df
```

```python
X=df.iloc[:,0:-1]
y=df['StressLevel']
```

```python
from sklearn.feature_selection import SelectKBest,mutual_info_classif

mic=SelectKBest(score_func=mutual_info_classif,k=3)
```

```python
from sklearn.feature_selection import SelectKBest,mutual_info_regression

mic=SelectKBest(score_func=mutual_info_regression,k=3)
mic.fit(X,y)
feature_MI_score=pd.Series(mic.scores_,index=X.columns)
feature_MI_score.sort_values(ascending=False)
```

```python
df=pd.read_csv('/content/mental_health_workplace_survey.csv')
df.head(5)
```

```python
df.columns
```

```python
df.drop(df[['EmployeeID']],axis=1)
```

```python
df.isna().sum()
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head(5)
```

```python
X=df.drop(['StressLevel','EmployeeID'],axis=1)
y=df['StressLevel']
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(train_X,train_y)
```

```python
reg.score(test_X,test_y)
```

```python
reg.score(train_X,train_y)
```

```python
from sklearn import linear_model
lasso_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(train_X,train_y)
```

```python
lasso_reg.score(test_X,test_y)
```

```python
lasso_reg.score(train_X,train_y)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(train_X,train_y)
```

```python
ridge_reg.score(test_X,test_y)
```

```python
ridge_reg.score(train_X,train_y)
```
