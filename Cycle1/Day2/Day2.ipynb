Selecting features using Mutual Information

```python
df.info()
```

```python
df.columns
```

```python
from sklearn.feature_selection import SelectKBest,mutual_info_classif
```

```python
df=pd.read_csv('/content/processed_data.csv')
```

```python
df.columns
```

```python
ab=df['SalaryRange']
```

```python
df.drop('SalaryRange',axis=1,inplace=True)
```

```python
df=pd.concat([df, ab],axis=1)
```

```python
df
```

```python
X=df.iloc[:,0:-1]
y=df['SalaryRange']
```

```python
mic=SelectKBest(score_func=mutual_info_classif,k=3)
```

```python
mic.fit(X,y)
feature_MI_score=pd.Series(mic.scores_,index=X.columns)
feature_MI_score.sort_values(ascending=False)
```
