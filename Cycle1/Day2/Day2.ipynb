```python
import pandas as pd
```

```python
df=pd.read_csv('/content/mental_health_workplace_survey.csv')
df.head(5)
```

```python
df.columns
```

```python
df2=df.copy()
```

```python
df2["Gender"].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
ohetransform=ohe.fit_transform(df2[["Gender"]])
```

```python
ohetransform
```

```python
df2=pd.concat([df, ohetransform],axis=1).drop(columns= ["Gender"])
```

```python
df2
```

```python
ohetransform=ohe.fit_transform(df2[["Country"]])
```

```python
ohetransform
```

```python
df2=pd.concat([df, ohetransform],axis=1).drop(columns= ["Country"])
```

```python
SalaryRanges=df3["SalaryRange"].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=[SalaryRanges])
```

```python
oe.fit_transform(df3[['SalaryRange']])
```

```python
df3['SalaryRange']=oe.fit_transform(df3[['SalaryRange']])
df3
```
