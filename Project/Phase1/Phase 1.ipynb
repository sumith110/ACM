```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df[~((df[numerical_cols] < (Q1 - 1.5 * IQR)) | (df[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

print("Shape of DataFrame before outlier removal:", df.shape)
print("Shape of DataFrame after outlier removal:", df_cleaned.shape)
```

```python
def generate_dropout_label(row):
    before = row['Mood Before Workout']
    after = row['Mood After Workout']
    intensity = row['Workout Intensity']
    duration = row['Workout Duration (mins)']

    duration_threshold = 45

    if (
        before in ['Tired', 'Stressed'] and
        after in ['Fatigued','Neutral'] and
        intensity in ['Low'] and
        duration < duration_threshold
    ):
        return 1  # Dropped out
    return 0  # Active
```

```python
df['Will drop'] = df.apply(generate_dropout_label, axis=1)
display(df.head())
```

```python
import torch
```

```python
import torch.nn as nn
```

```python
import torch.nn.functional as F
```

```python
ab.shape
```

```python
class Model(nn.Module):
   def __init__(self,in_features=32,h1=8,h2=8,out_features=2):
    super().__init__()
    self.fc1=nn.Linear(in_features,h1)
    self.fc2=nn.Linear(h1,h2)
    self.fc3=nn.Linear(h2,out_features)
   def forward(self,x):
    x=F.relu(self.fc1(x))
    x=F.relu(self.fc2(x))
    x=self.fc3(x) # Corrected from self.out(x)

    return x
```

```python
torch.manual_seed(41)
```

```python
model=Model()
```

```python
X=ab.drop('Will drop',axis=1)
```

```python
y=ab['Will drop']
```

```python
X=X.values
y=y.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train,X_Test,y_Train,y_Test=train_test_split(X,y,test_size=0.2)
```

```python
X_Train=torch.FloatTensor(X_Train)
X_Test=torch.FloatTensor(X_Test)
```

```python
y_Train=torch.Tensor(y_Train)
y_Test=torch.Tensor(y_Test)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer =torch.optim.Adam(model.parameters(),lr=0.1)
```

```python
model.parameters
```

```python
# Convert labels to long for CrossEntropyLoss
y_Train = y_Train.long()
y_Test = y_Test.long()

epochs = 100
losses = []

for i in range(epochs):
    y_pred = model(X_Train)  # forward pass
    loss = criterion(y_pred, y_Train)  # no parentheses, use long() labels

    losses.append(loss.item())

    if i % 10 == 0:
        print(f'Epoch: {i}  Loss: {loss.item():.4f}')

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
with torch.no_grad():
  y_eval = model.forward(X_Test)
  loss=criterion(y_eval,y_Test)
```

```python
loss
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

# Get predictions
_, predicted = torch.max(y_eval, 1)

# Convert tensors to numpy arrays for sklearn metrics
predicted_np = predicted.numpy()
y_Test_np = y_Test.numpy()

# Calculate metrics
accuracy = accuracy_score(y_Test_np, predicted_np)
f1 = f1_score(y_Test_np, predicted_np)
precision = precision_score(y_Test_np, predicted_np)
recall = recall_score(y_Test_np, predicted_np)

print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
print(f'Precision: {precision:.4f}')
print(f'Recall: {recall:.4f}')
```
