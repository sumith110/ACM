```python
from sklearn.datasets import fetch_20newsgroups
```

```python
newsgroups = fetch_20newsgroups(subset='all')
X,y = newsgroups.data,newsgroups.target_names
```

```python
len(X)
```

```python
len(y)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_tfidf = vectorizer.fit_transform(X)
```

```python
X_tfidf.shape
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=2, random_state=42)
X_reduced = svd.fit_transform(X_tfidf)
```

```python
X_reduced.shape
```

```python
import matplotlib.pyplot as plt
```

```python
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y)
plt.xlabel("Comp 1")
plt.ylabel("Comp 2")
```

```python
from sklearn.cluster import KMeans
```

```python
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y)
plt.title(" Clusters")
plt.xlabel("Comp 1")
plt.ylabel("Comp 2")
```

```python
kmeans = KMeans(n_clusters=20, random_state=42)
y_pred = kmeans.fit_predict(X_reduced)
```

```python
scatter2 = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_pred)
plt.title("K-Means Predicted Clusters")
plt.xlabel("Comp 1")
plt.ylabel("Comp 2")
```

```python
from sklearn.metrics import silhouette_score
```

```python
score = silhouette_score(X_reduced, y_pred)
score
```
