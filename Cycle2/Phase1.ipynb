```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['B','M']])

```

```python
df['diagnosis']=oe.fit_transform(df[['diagnosis']])
ab=df['diagnosis']
```

```python
X=df.drop('diagnosis',axis=1)
```

```python
df
```

```python
X = df.drop(['diagnosis','id'],axis=1)
y = df['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaled_df = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,stratify=y, random_state=42)
```

#DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
dtc.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(dtc,X,y,cv=5)
scores
```

```python
scores.mean()
```

#Bagging

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bagmodel=BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.7,
    random_state=0
)
```

```python
bagmodel.fit(X_train,y_train)
```

```python
bagmodel.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(bagmodel,X,y,cv=5)
scores
```

```python
scores.mean()
```

#AdaBoosting

```python
from sklearn.ensemble import AdaBoostClassifier
ada_boost = AdaBoostClassifier(n_estimators=100, random_state=0)
```

```python
ada_boost.fit(X_train, y_train)
ada_boost.score(X_test,y_test)
```

```python
scores = cross_val_score(ada_boost, X, y, cv=5)
scores.mean()
```

#XGBoosting


```python
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators=100, random_state=0)
```

```python
xgb.fit(X_train, y_train)
```

```python
xgb.score(X_test, y_test)
```

```python
scores = cross_val_score(xgb, X, y, cv=5)
scores.mean()
```
